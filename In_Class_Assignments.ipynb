```python
sales = float(input(f"Enter Sales ($): "))
print(f"Sales: ${sales}")
COGS = sales * 0.4
print(f"Cost of Goods Sold: ${COGS}")
Gross_Profit = sales - COGS
print(f"Gross Profit: ${Gross_Profit}")

Tax_Rate = float(input("Enter Tax Rate (%): "))
Taxes = Gross_Profit * (Tax_Rate/100)
Net_Income = Gross_Profit - Taxes

print(f"Taxes: ${Taxes}")
print(f"Net Income: ${Net_Income}")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input ("Choose numerical operation (e.g., +, -, *, etc.): ")

res = None

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op =="//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```

```python
print("Hour   Fee")

hour = 1
flat_fee = 5
rate = 2.5

while hour < 9:
  cost = flat_fee + (rate*hour)
  if cost <10:
    cost = 10.0
  elif cost > 20:
    cost = 20.0
  print(f" {hour}  =  {cost}")
  hour += 1
```

```python
fixedcost = 200
ticketprice = 10
att = 20

print("$Ad\tProfit")
for adcost in range(0, 225, 25):
  addatt = 2 * ((adcost) ** 0.5)
  totalcost = fixedcost + adcost
  gross = (ticketprice * att) + (round(addatt) * ticketprice)
  net = gross - totalcost
  print(f" {float(adcost)}\t{float(net)}")
```

```python
import random
check = input("Do you want to play (y/n): ").lower()

player_score = 0
computer_score = 0

while check == "y":
  player_move = input("Choose rock, paper, or scissors: ").lower()
  computer_move = random.choice(['rock','paper','scissors'])
  print(f"The computer chose {computer_move}")
  if computer_move == player_move:
    print("tie")
  elif (
      (computer_move == "paper" and player_move == "rock") or
      (computer_move == "rock" and player_move == "scissors") or
      (computer_move == "scissors" and player_move == "paper")
      ):
    print("Computer Wins!!")
    computer_score += 1
    print(f"Score: Player {player_score}  Computer {computer_score}")
    check = input("Do you want to play again (y/n): ")
  elif (
      (player_move == "paper" and computer_move == "rock") or
      (player_move == "rock" and computer_move == "scissors") or
      (player_move == "scissors" and computer_move == "paper")
      ):
    print("Player Wins")
    player_score += 1
    print(f"Score: Player {player_score}  Computer {computer_score}")
    check = input("Do you want to play again (y/n): ")
else:
  print("Thank you for playing!")
  print(f"Final Score: Player {player_score}  Computer {computer_score}")
```
