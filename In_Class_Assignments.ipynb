```python
sales = float(input(f"Enter Sales ($): "))
print(f"Sales: ${sales}")
COGS = sales * 0.4
print(f"Cost of Goods Sold: ${COGS}")
Gross_Profit = sales - COGS
print(f"Gross Profit: ${Gross_Profit}")

Tax_Rate = float(input("Enter Tax Rate (%): "))
Taxes = Gross_Profit * (Tax_Rate/100)
Net_Income = Gross_Profit - Taxes

print(f"Taxes: ${Taxes}")
print(f"Net Income: ${Net_Income}")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input ("Choose numerical operation (e.g., +, -, *, etc.): ")

res = None

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op =="//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```

```python
print("Hour   Fee")

hour = 1
flat_fee = 5
rate = 2.5

while hour < 9:
  cost = flat_fee + (rate*hour)
  if cost <10:
    cost = 10.0
  elif cost > 20:
    cost = 20.0
  print(f" {hour}  =  {cost}")
  hour += 1
```

```python
fixedcost = 200
ticketprice = 10
att = 20

print("$Ad\tProfit")
for adcost in range(0, 225, 25):
  addatt = 2 * ((adcost) ** 0.5)
  totalcost = fixedcost + adcost
  gross = (ticketprice * att) + (round(addatt) * ticketprice)
  net = gross - totalcost
  print(f" {float(adcost)}\t{float(net)}")
```
