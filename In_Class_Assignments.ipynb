```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input ("Choose numerical operation (e.g., +, -, *, etc.): ")

res = None

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op =="//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```

```python
import random
check = input("Do you want to play (y/n): ").lower()

player_score = 0
computer_score = 0

while check == "y":
  player_move = input("Choose rock, paper, or scissors: ").lower()
  computer_move = random.choice(['rock','paper','scissors'])
  print(f"The computer chose {computer_move}")
  if computer_move == player_move:
    print("tie")
  elif (
      (computer_move == "paper" and player_move == "rock") or
      (computer_move == "rock" and player_move == "scissors") or
      (computer_move == "scissors" and player_move == "paper")
      ):
    print("Computer Wins!!")
    computer_score += 1
    print(f"Score: Player {player_score}  Computer {computer_score}")
    check = input("Do you want to play again (y/n): ")
  elif (
      (player_move == "paper" and computer_move == "rock") or
      (player_move == "rock" and computer_move == "scissors") or
      (player_move == "scissors" and computer_move == "paper")
      ):
    print("Player Wins")
    player_score += 1
    print(f"Score: Player {player_score}  Computer {computer_score}")
    check = input("Do you want to play again (y/n): ")
else:
  print("Thank you for playing!")
  print(f"Final Score: Player {player_score}  Computer {computer_score}")
```

```python
def loan_payment(principal, annual_rate, months):
  assert (isinstance(principal, (int, float)) and
        isinstance(annual_rate, (int, float)) and
        isinstance(months, (int, float)))
  if (principal <= 0) or (annual_rate) < 0 or (months <= 0):
    raise ValueError("Invalid input")

  try:
    r = annual_rate/12
    if r == 0:
        pmt = principal/months
    else:
      pmt = r * principal / (1 - (1 + r) ** (-months))
  except Exception as e:
    print(f"Error: {e}")

loan_payment(1500, 120, 13)
```

```python
class Greeting:
  def __init__(self, first_name, last_name):
    self.first_name = first_name
    self.last_name = last_name

  def display(self, formal=False):
    if formal:
      print(f"{self.last_name}, {self.first_name}")
    else:
      print(self.first_name)

  def __str__(self):
    return f"Instance of class 'Greeting'"

name = Greeting("John", "Deer")
print(name)
name.display()
name.display(formal=True)
```

```python
class Student:
  def __init__(self, name, student_id):
    self.__name = name
    self.student_id = student_id
    self.__courses = []

  def get_name(self):
    return self.__name

  def get_student_id(self):
    return self.student_id

  def get_courses(self):
    return self.__courses

  def set_name(self, newName):
    assert isinstance(newName, str)
    self.__name = newName

  def courses_count(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
      print(f"{self.__name} enrolled in {newCourse}")
    else:
      print(f"{self.__name} is already enrolled in {newCourse}")
```

```python
s = Student("John Smith", 12345)
print(f"Student name: {s.get_name()}")
print(f"Show currently enrolled courses: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Show currently enrolled courses: {s.get_courses()}")
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    return (dx + dy) ** 0.5


class Rectangle:
  def __init__(self, p1, p2, p3, p4):
    self.p1 = p1
    self.p2 = p2
    self.p3 = p3
    self.p4 = p4
    self.width = self.width_()
    self.height = self.height_()

  def width_(self):
    return self.p1.distance(self.p2)

  def height_(self):
    return self.p1.distance(self.p3)

  def area(self):
    return self.width * self.height



p1 = Point(0,0)
p2 = Point(5,0)
p3 = Point(0,5)
p4 = Point(5,5)


z = Rectangle(p1, p2, p3, p4)
z.area()
```

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.__wage = wage
    self.hours = hours

  def calculate__pay(self):
    return self.__wage * self.hours

  def __str__(self):
    return f"Employee Name: {self.name}"

  def get_wage(self):
    return self.__wage()

  def set_wage(self, NewWage):
    assert NewWage > 0, "Wage valyes cannot be negative or zero"
    self.__wage = NewWage

e = Employee('John Smith', 15, 10)
e.calculate__pay()
```
