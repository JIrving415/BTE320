```python
sales = float(input(f"Enter Sales ($): "))
print(f"Sales: ${sales}")
COGS = sales * 0.4
print(f"Cost of Goods Sold: ${COGS}")
Gross_Profit = sales - COGS
print(f"Gross Profit: ${Gross_Profit}")

Tax_Rate = float(input("Enter Tax Rate (%): "))
Taxes = Gross_Profit * (Tax_Rate/100)
Net_Income = Gross_Profit - Taxes

print(f"Taxes: ${Taxes}")
print(f"Net Income: ${Net_Income}")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input ("Choose numerical operation (e.g., +, -, *, etc.): ")

res = None

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op =="//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```
